```python
from langchain_community.chat_models import ChatOllama
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage 
from langchain_core.prompts import ChatPromptTemplate as cpt
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
import os 
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI as genai 
```

```python
from langchain_core.runnables import RunnableParallel,RunnableLambda
```

```python
#paller chain 1
l_prompt=cpt.from_messages([
        ("system","You are a social media post generator"),
        ("human","creat a post for the following text for  linkdin: {text}"),
    ])

#task 2 
llm=ChatOllama(model="deepseek-r1:1.5b")
llm1=genai(model="gemini-2.5-flash")

#task3
str_parser=StrOutputParser()

cahin_linkdin=l_prompt| llm1 |str_parser
```

```python
#plaler chain 2 
def insta_chian(text:dict):
    text=text["text"]
    
    i_prompt=cpt.from_messages([
        ("system","You are a social media post generator"),
        ("human","creat a post for the following text for  instagram: {text}"),
    ])

    #task 2 
    llm=ChatOllama(model="deepseek-r1:1.5b")
    llm1=genai(model="gemini-2.5-flash")

    #task3
    str_parser=StrOutputParser()

    chain_insta=i_prompt | llm1 | str_parser
    res=chain_insta.invoke(text)
    return res
insta_cahin_runable=RunnableLambda(insta_chian)
```

```python
from langchain_core.runnables import RunnableParallel
#final orchestration

final_chain = RunnableParallel(
    linkdin=cahin_linkdin,
    instagram=insta_cahin_runable
)
```
